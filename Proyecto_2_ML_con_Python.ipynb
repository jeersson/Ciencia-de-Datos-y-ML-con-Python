Generar un Notebook Colab con código y resultados, que contiene
información de autos de un periodo de tiempo. Resolver los siguientes:
- Limpieza y tratamiento de las variables (Según corresponda)
- Generar 5 nuevas variables aplicando los siguientes criterios con las
variables numéricas:

1. Log10(var1/var2)
2. Sqrt(var1)*exp(var2)/200

3. Si var 1 > var 2 → 5, caso contrario 3
4. 1/logn(var1/var2)*100
5. var2**2/var1

```python
# importamos las librerías
import pandas as pd
import math
```

```python
#levantamos la data con pandas
data_autos = pd.read_csv('autos.csv', sep=';')
data_autos.head()
```

```python
def logaritmo(var1, var2):
  return math.log10(var1/var2)
```

```python
def raiz(var1, var2):
  return math.sqrt(var1)**(var2)/200
```

```python
def condicional(var1, var2):
  if(var1 > var2):
    return 5
  else:
    return 3
```

```python
def logN(var1, var2):
  k = 1
  if(math.log(var1/var2)*100 != 0):
    k = math.log(var1/var2)*100
  return 1/k
```

```python
def exponencial(var1, var2):
  return var2**2/var1
```

```python
data_autos['Log10'] = data_autos.apply(lambda x: logaritmo(x['Millas_Por_Galon'], x['Aceleracion']), axis=1)
data_autos['Raiz'] = data_autos.apply(lambda x: raiz(x['Millas_Por_Galon'], x['Aceleracion']), axis=1)
data_autos['Condicional'] = data_autos.apply(lambda x: condicional(x['Millas_Por_Galon'], x['Aceleracion']), axis=1)
data_autos['LogN'] = data_autos.apply(lambda x: logN(x['Millas_Por_Galon'], x['Aceleracion']), axis=1)
data_autos['Exponencial'] = data_autos.apply(lambda x: exponencial(x['Millas_Por_Galon'], x['Aceleracion']), axis=1)
data_autos.head()
```

**Ejercicio No 4**
- Crear una función que muestre los siguientes gráficos
(simultáneamente) para las variables numéricas, según una variable
cualitativa:

❖ Cajas y bigotes
❖ Barras de la media
❖ Histograma con densidad (Curva)

La función debe contar con su respectiva documentación.

```python
#cargamos las librerías correspondientes
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy import stats
import missingno as msno
```

```python
#partimos de esta data de ejemplo
df = pd.read_csv('autos.csv', sep=";") 
df.head()
```

```python
# creamos un df más pequeño de ejemplo
df2=df[['Pais_Origen','Anyo','Aceleracion', 'Cilindros']]
df2.head()
```

```python
def graficos_varios(data, x_var, y_var, group_var): 
    """
    Genera tres gráficos simultáneos para una variable numérica agrupada por una variable categórica:
    - Boxplot por grupo
    - Barplot de la media con intervalo de confianza
    - Histograma con densidad (curva) por grupo

    Parámetros
    ----------
    data : DataFrame
        Datos a graficar
    x_var : str
        Nombre de la variable categórica (eje X)
    y_var : str
        Nombre de la variable numérica (eje Y)
    group_var : str
        Nombre de la variable categórica para agrupar los datos y colorear los gráficos
    """

    # matriz de graficos
    fig, (ax_box, ax_bar, ax_hist) = plt.subplots(1, 3, figsize=(45, 10), sharex=True)

    # boxplot
    sns.boxplot(data=data, x=x_var, y=y_var, hue=group_var, ax=ax_box)

    # barplot
    sns.barplot(data=data, x=x_var, y=y_var, estimator = np.mean, hue=group_var, errorbar=('ci', 99), ax=ax_bar)

    # histograma
    sns.histplot(data=data, x=y_var, hue=group_var, element='step', kde=True, ax=ax_hist)

    # establecemos los axis y títulos
    ax_box.set_xlabel('')
    ax_box.set_ylabel(y_var)
    ax_bar.set_xlabel('')
    ax_bar.set_ylabel('Media')
    ax_hist.set_xlabel(y_var)
    ax_hist.set_ylabel('Densidad')
    ax_box.set_title('Cajas y bigotes')
    ax_bar.set_title('Barras')
    ax_hist.set_title('Histograma')

    plt.tight_layout()
    plt.show()
```

```python
graficos_varios(data=df2, x_var='Pais_Origen', y_var='Aceleracion', group_var='Pais_Origen')
```
